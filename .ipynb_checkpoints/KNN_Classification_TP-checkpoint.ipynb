```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
nom_du_fichier_csv = "Social_Network_Ads.csv"
chemin_complet = "C:\\Users\\hil\\Documents\\ilyas\\5IIR14\\system info geo\\TP1\\" + nom_du_fichier_csv
dataset = pd.read_csv(chemin_complet)
print(dataset.head())
```

```python
from sklearn.preprocessing import LabelEncoder
dataset['Gender']= LabelEncoder().fit_transform(dataset['Gender'].astype(str))
X = dataset.iloc[:, [1, 3]].values
Y = dataset.iloc[:, 4].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)

X_test = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(y_pred)
```

```python
c=0
for i in range(0,len(y_pred)):
    if(y_pred[i]==y_test[i]):
        c=c+1
        accuracy=c/len(y_pred)
print("Accuracy is")
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
matrice_conf=pd.DataFrame(cm,index=['positive','négative'],columns=['positive','negative'])
```
